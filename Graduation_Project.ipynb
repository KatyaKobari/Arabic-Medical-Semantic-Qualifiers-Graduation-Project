```python
class ArabicSymptomNER(nn.Module):
    def __init__(self, model_name, num_labels, hidden_dim=256, dropout_prob=0.1):
        super().__init__()
        self.bert = AutoModel.from_pretrained(model_name)
        self.lstm = nn.LSTM(
            input_size=self.bert.config.hidden_size,
            hidden_size=hidden_dim,
            num_layers=1,
            bidirectional=True,
            batch_first=True
        )
        self.dropout = nn.Dropout(dropout_prob)
        self.classifier = nn.Linear(hidden_dim * 2, num_labels)
        self.crf = CRF(num_labels)

    def forward(self, input_ids, attention_mask, labels=None):
        bert_out = self.bert(
            input_ids=input_ids,
            attention_mask=attention_mask
        ).last_hidden_state

        lstm_out, _ = self.lstm(bert_out)

        emissions = self.classifier(self.dropout(lstm_out))

        if labels is not None:
            mask_bool = attention_mask.bool()
            loss = -self.crf(emissions, labels, mask=mask_bool)
            return loss
        else:
            mask_bool = attention_mask.bool()
            predictions = self.crf.viterbi_decode(emissions, mask=mask_bool)
            return predictions

# model = ArabicSymptomNER(model_name, len(tag2id))
# optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)
# device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# model.to(device)
```

```python
def extract_predictions_from_bio(Parsed_path, Bio_prediction, Parsed_Models):
    with open(Parsed_path, 'r', encoding='utf-8') as f:
        parsed_data = json.load(f)

    with open(Bio_prediction, 'r', encoding='utf-8') as f:
        lines = [line.strip() for line in f]

    dialogue_bio_blocks = []
    current_block = []
    for line in lines:
        if line == "":
            if current_block:
                dialogue_bio_blocks.append(current_block)
                current_block = []
        else:
            current_block.append(line)
    if current_block:
        dialogue_bio_blocks.append(current_block)

    assert len(dialogue_bio_blocks) == len(parsed_data), \
        f"Mismatch: {len(dialogue_bio_blocks)} BIO dialogues vs {len(parsed_data)} parsed dialogues."

    for idx, bio_block in enumerate(dialogue_bio_blocks):
        new_symptoms = process_bio_dialogue(bio_block)
        parsed_data[idx]["Symptoms"] = merge_duplicate_symptoms(new_symptoms)

    with open(Parsed_Models, 'w', encoding='utf-8') as f:
        json.dump(parsed_data, f, ensure_ascii=False, indent=2)


def process_bio_dialogue(bio_lines):
        tokens, tags = [], []
        for line in bio_lines:
            try:
                word, tag = line.split()
                if word.startswith("##") and tokens:
                    tokens[-1] += word[2:]
                    if tags[-1] == "O" and tag != "O":
                        tags[-1] = tag
                else:
                    tokens.append(word)
                    tags.append(tag)
            except ValueError:
                continue

        symptoms = []
        i = 0
        current_symptom = None
        pending_location = ""

        while i < len(tokens):
            tag = tags[i]
            token = tokens[i]

            if tag == 'B-SYMPTOM' or tag == 'I-SYMPTOM':
                symptom_phrase = [token]
                j = i + 1
                while j < len(tokens) and tags[j] == 'I-SYMPTOM':
                    symptom_phrase.append(tokens[j])
                    j += 1
                symptom_text = " ".join(symptom_phrase)
                current_symptom = {
                    "name": symptom_text,
                    "location": pending_location or "",
                    "time": [],
                    "duration": "",
                    "severity": "",
                    "spread_to": "",
                    "original_phrase": symptom_text
                }
                symptoms.append(current_symptom)
                pending_location = ""
                i = j
                continue

            if not current_symptom:
                i += 1
                continue

            if tag == 'B-LOCATION':
                location_phrase = [token]
                j = i + 1
                while j < len(tokens) and tags[j] == 'I-LOCATION':
                    location_phrase.append(tokens[j])
                    j += 1
                location_text = " ".join(location_phrase)

                if j < len(tokens) and (tags[j] == 'B-SYMPTOM' or tags[j] == 'I-SYMPTOM'):
                    pending_location = location_text
                else:
                    current_symptom["location"] = location_text
                i = j
                continue

            if tag == 'B-TIME':
                time_phrase = [token]
                j = i + 1
                while j < len(tokens) and tags[j] == 'I-TIME':
                    time_phrase.append(tokens[j])
                    j += 1
                current_symptom["time"].append(" ".join(time_phrase))
                i = j
                continue

            if tag == 'B-DURATION':
                duration_phrase = [token]
                j = i + 1
                while j < len(tokens) and tags[j] == 'I-DURATION':
                    duration_phrase.append(tokens[j])
                    j += 1
                current_symptom["duration"] = " ".join(duration_phrase)
                i = j
                continue

            if tag == 'B-SEVERITY':
                severity_phrase = [token]
                j = i + 1
                while j < len(tokens) and tags[j] == 'I-SEVERITY':
                    severity_phrase.append(tokens[j])
                    j += 1
                current_symptom["severity"] = " ".join(severity_phrase)
                i = j
                continue

            if tag == 'B-SPREAD':
                spread_phrase = [token]
                j = i + 1
                while j < len(tokens) and tags[j] == 'I-SPREAD':
                    spread_phrase.append(tokens[j])
                    j += 1
                current_symptom["spread_to"] = " ".join(spread_phrase)
                i = j
                continue

            i += 1
        return symptoms
```

```python
def merge_duplicate_symptoms(symptoms):
    grouped = {}
    for sym in symptoms:
        name = sym.get("name", "").strip()
        if not name:
            continue
        if name not in grouped:
            grouped[name] = sym.copy()
            if isinstance(grouped[name].get("time"), str):
                grouped[name]["time"] = [grouped[name]["time"]]
            elif grouped[name].get("time") is None:
                grouped[name]["time"] = []
        else:
            existing = grouped[name]
            if not existing.get("location") and sym.get("location"):
                existing["location"] = sym["location"]
                if sym.get("location_SNOMED"):
                    existing["location_SNOMED"] = sym["location_SNOMED"]
            if not existing.get("severity") and sym.get("severity"):
                existing["severity"] = sym["severity"]
                if sym.get("severity_SNOMED"):
                    existing["severity_SNOMED"] = sym["severity_SNOMED"]
            if not existing.get("duration") and sym.get("duration"):
                existing["duration"] = sym["duration"]
                if sym.get("duration_SNOMED"):
                    existing["duration_SNOMED"] = sym["duration_SNOMED"]
            if not existing.get("spread_to") and sym.get("spread_to"):
                existing["spread_to"] = sym["spread_to"]
            if sym.get("time"):
                if not existing.get("time"):
                    existing["time"] = []
                for t in sym["time"]:
                    if t not in existing["time"]:
                        existing["time"].append(t)
    return list(grouped.values())
```

```python
def diagnosis_classification(Parsed_snomed,classification_diagnoses):

    model = joblib.load(LR_model)
    symptom_list = joblib.load(combined_symptom_list)

    mapping_df = pd.read_csv(Standardized_Symptoms, encoding="utf-8")
    arabic_to_english = dict(zip(mapping_df["Standard Arabic"], mapping_df["English Symptom"]))

    with open(Parsed_snomed, "r", encoding="utf-8") as f:
        parsed_data = json.load(f)

    results = []

    for dialogue in parsed_data:
        symptoms = dialogue["Symptoms"]

        mapped_symptoms = set()
        all_symptom_entries = []

        for sym in symptoms:
            name = sym.get("name", "").strip()
            location = sym.get("location", "").strip()

            if name in location_ignored_symptoms or not location:
                key = name
            else:
                key = f"{name} {location}"

            english = arabic_to_english.get(key) or arabic_to_english.get(name)

            if english:
                if english in symptom_list:
                    mapped_symptoms.add(english)
                all_symptom_entries.append({
                    "original": key,
                    "mapped_to": english,
                    "mapped": True
                })
            else:
                all_symptom_entries.append({
                    "original": key,
                    "mapped_to": None,
                    "mapped": False
                })

        input_vector = [1 if symptom in mapped_symptoms else 0 for symptom in symptom_list]
        input_vector = np.array(input_vector).reshape(1, -1)

        probs = model.predict_proba(input_vector)[0]
        classes = model.classes_

        top_idx = np.argmax(probs)
        top_diagnosis = classes[top_idx]
        confidence = round(probs[top_idx] * 100, 2)

        if top_diagnosis in disease_dict:
            diagnosis_en = top_diagnosis + f" ({confidence}%)"
            diagnosis_ar = disease_dict[top_diagnosis]
        else:
            diagnosis_en = top_diagnosis + f" ({confidence}%)"
            diagnosis_ar = "لم أستطع التشخيص – يُرجى مراجعة الطبيب"

        results.append({
            "Symptoms_mapping": all_symptom_entries,
            "Mapped_symptoms_for_model": list(mapped_symptoms),
            "Predicted_diagnosis_en": diagnosis_en,
            "Predicted_diagnosis_ar": diagnosis_ar
        })

    with open(classification_diagnoses, "w", encoding="utf-8") as f:
        json.dump(results, f, ensure_ascii=False, indent=2)
```

```python
map_location=torch.device('cpu')
translator = Translator()
app = Flask(__name__)
CORS(app)

SYMPTOM_TRANSLATE = {}
try:
    df_sym = pd.read_csv(SymptomMap)
    SYMPTOM_TRANSLATE = dict(zip(df_sym["Arabic Symptom"],
                                 df_sym["English Term"]))
except Exception as e:
    print("⚠  مشكلة في قراءة symptom_mapping.csv:", e)


@app.route("/", methods=["GET"])
def page1():
    return render_template_string(form_page)

@app.route("/final_submit", methods=["POST"])
def final_submit():
    try:
        req = request.json
        lang = req.get("lang", "ar")
        text = req.get("text", "")
        age = req.get("age")
        gender = req.get("gender")
        medical = req.get("medicalHistory")
        family = req.get("familyHistory")
        smoking = req.get("smoking")

        parsed = parse_dialogue(text, age, gender, medical, family, smoking)

        with open(Parsed_path, "w", encoding="utf-8") as f:
            json.dump([parsed], f, ensure_ascii=False, indent=2)

        predict_bio_tags(Parsed_path, Bio_prediction)
        extract_predictions_from_bio(Parsed_path, Bio_prediction, Parsed_Models)
        normalize_symptoms_with_snomed(Parsed_Models, Parsed_snomed)


        # ✅ تحميل ودمج الأعراض
        with open(Parsed_snomed, "r", encoding="utf-8") as f:
            data = json.load(f)

        if data and isinstance(data, list):
            symptoms = data[0].get("Symptoms", [])
            data[0]["Symptoms"] = merge_duplicate_symptoms(symptoms)

            with open(Parsed_snomed, "w", encoding="utf-8") as f:
                json.dump(data, f, ensure_ascii=False, indent=2)

        diagnosis_classification(Parsed_snomed, classification_diagnoses)

        with open(classification_diagnoses, "r", encoding="utf-8") as f:
            diag = json.load(f)[0]

        diag_en = diag.get("Predicted_diagnosis_en", "")
        diag_ar = diag.get("Predicted_diagnosis_ar", "")

        def match(text, search_keys=True):
            items = disease_dict.keys() if search_keys else disease_dict.values()
            for item in items:
                if item.lower() in text.lower():
                    return item
            return None

        diagnosis = match(diag_en if lang == "en" else diag_ar, lang == "en")
        if not diagnosis:
            diagnosis = "Diagnosis could not be determined" if lang == "en" else "لم يتم التشخيص"

        return jsonify({"diagnosis": diagnosis})

    except Exception as e:
        print("❌ Error in final_submit:", e)
        err_msg = "حدث خطأ أثناء التحليل" if lang == "ar" else "Error during analysis"
        return jsonify({"diagnosis": err_msg}), 500

@app.route("/analyze_data", methods=["POST"])
def analyze_data():
    """
    يستقبل JSON فيه {"lang": "ar" أو "en"}
    ويعيد السجل مترجماً عند الحاجة.
    يبقي Full_dialogue و original_phrase بالعربية دائماً.
    """
    try:
        lang = request.json.get("lang", "ar")   # افتراضي عربي

        with open(Parsed_snomed, "r", encoding="utf-8") as f:
            data = json.load(f)
        if not data:
            return jsonify({})

        record = data[0]      # نعالج أول حالة فقط

        # —— الترجمة إلى الإنجليزية عند الطلب ——
        if lang == "en":
            protected_top = {"Full_dialogue"}      # تظل عربي
            protected_sym = {"original_phrase"}    # تظل عربي

            # 1) ترجمة الحقول العلوية
            for k, v in record.items():
                if k not in protected_top and isinstance(v, str) and v.strip():
                    record[k] = translator.translate(v, dest="en").text

            # 2) ترجمة كل عَرَض داخل Symptoms
            for sym in record.get("Symptoms", []):
                # أ) اسم العَرَض
                if isinstance(sym.get("name"), str) and sym["name"].strip():
                    ar_sym = sym["name"]
                    sym["name"] = SYMPTOM_TRANSLATE.get(
                        ar_sym,
                        translator.translate(ar_sym, dest="en").text
                    )

                # ب) الموقع
                if isinstance(sym.get("location"), str) and sym["location"].strip():
                    ar_loc = sym["location"]
                    sym["location"] = LocationTranslate.get(
                        ar_loc,
                        translator.translate(ar_loc, dest="en").text
                    )

                # ج) باقي الحقول النصية
                for k, v in sym.items():
                    if k in protected_sym or k in {"name", "location"}:
                        continue
                    if isinstance(v, str) and v.strip():
                        sym[k] = translator.translate(v, dest="en").text

                # د) time لو كانت قائمة
                if isinstance(sym.get("time"), list):
                    sym["time"] = [
                        translator.translate(t, dest="en").text for t in sym["time"]
                    ]

        return jsonify(record)

    except Exception as e:
        print("❌ Error loading analysis:", e)
        return jsonify({})

# ---------------- تشغيل التطبيق ----------------
public_url = ngrok.connect(5000)
print("🔥 App is running on:", public_url)
app.run(port=5000)
```
